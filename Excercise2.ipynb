```python
#class count
count_class_0,count_class_1 = df['class'].value_counts()
print(count_class_0)
print(count_class_1)
```

```python
#Divide by class
x = df[df['class'] == 0]
print(x)
```

```python
y = df[df['class'] == 0]
print(y)
```

```python
df_class_0 = df[df['class'] == 0]
print(df_class_0)
```

```python
df_class_1 = df[df['class'] == 1]
print(df_class_1)
```

```python
#Undersample 0-class and concat the DataFrame of both class
a = df_class_1.sample(count_class_1)
df_test_under = pd.concat([a, df_class_0],axis = 0)

print("Random under-sampling:")
print(df_test_under['class'].value_counts())
```

```python
print('df_class_0 size:',df_class_0.size)
print('count_class_0 size:',count_class_0)
```

```python
print('df_class_1 size:',df_class_1.size)
print('count_class_1 size:',count_class_1)
```

```python
#Oversample 1-class and concat the DataFrame of both classes
df_class_1_over = df_class_0.sample(count_class_0,replace = True)
df_test_over = pd.concat([df_class_1,df_class_1_over],axis = 0)

print("Random over-sampling:")
print(df_test_over['class'].value_counts())
```
