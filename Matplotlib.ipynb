```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Graph

```python
#basic graph
m = [0,2,4,6,8]
plt.figure(figsize = (12,5), dpi = 100) #dpi - dot per inches(Resolution of graphs)
plt.plot(m)
```

```python
#Keyword Argument Notation
plt.plot(m,color='pink',linewidth=2,
         marker='x',linestyle='--',markersize=10,
         markeredgecolor='blue',markerfacecolor='orange')
plt.show()
```

```python
#Shorthand Notation
#fmt = '[color][marker][line]'
plt.plot(m,'r.-')
plt.show()
```

```python
#Temparature in a week
a = [1,2,3,4,5,6,7]
b = [36,37,38,40,39,37,38]

fig = plt.figure(figsize = (12,5),dpi = 100)

plt.plot(a,b,color='pink',linewidth=2,
         marker='x',linestyle='--',markersize=10,
         markeredgecolor='blue',markerfacecolor='orange')
plt.title("Temparature in a week")
plt.xlabel('Days')
plt.ylabel('Temparature')

plt.savefig('t3.png',dpi = 300)#should be before plt.show;else blank pic
plt.show()
```

```python
x1=[1,16,30,42,55,68,77,88]
y1=[1,4,9,16,25,36,49,64]

x2=[1,6,12,18,28,40,52,65]
y2=[2,6,8,10,12,24,32,73]

fig=plt.figure()

l1=plt.plot(x1,y1,'rD-.')#solid line with yellow color and square marker
l2=plt.plot(x2,y2,'gs:')#dash line with green color and circle marker
plt.legend(['tv','smartphone'],loc='right')#legend placed at lower right
plt.title('Advertisement effects on sales')
plt.xlabel('medium')
plt.ylabel('sales')
plt.show()
```

```python
import numpy as np
#select interval we want to plot points at
x = np.arange(0,4.5,0.5) #0,0.5,1,1.5,2,2.5,3,3.5,4
print(x)
#plot part of the graph as line
plt.plot(x[:6],x[:6]**2,'r^-') #0,0.5,1,1.5,2,2.5
#plot remainder of graph as dot
plt.plot(x[5:],x[5:]**2,'g^--') #2.5,3.0,3.5,4.0
#Add a title(specify font parameters with fontdict)
plt.title('Our First Graph!',
        fontdict = {'name' : 'Monotype Corsiva',
                    'size' : 30,
                    'color' : 'red',
                    'weight' : 'bold'
                   })
#X and Y labels
plt.xlabel('X Axis',fontdict = {'name' : 'Monotype Corsiva',
                    'size' : 20,
                    'color' : 'yellow',
                    'weight' : 'bold',
                               })
plt.xlabel('Y Axis',fontdict = {'name' : 'Monotype Corsiva',
                    'size' : 20,
                    'color' : 'yellow',
                    'weight' : 'bold',
                               })
#X, Y axis Tickmarks(scale your graph)
plt.grid(color = 'b',linestyle = ':',linewidth = 1)
plt.xticks([0,1,2,3,4,5]) #Important
plt.yticks([0,4,8,12,16])
```

### Bar Graph

```python
categories_bar = ['category A','category B','category C','category D']
values_bar=[25,40,30,20]

plt.figure(figsize=(10,5))
plt.bar(categories_bar,values_bar,color=['blue','green','yellow','red'])
plt.title('bar graph')
plt.xlabel('categories')
plt.ylabel('values')
plt.show()
```

### Pie chart

```python
categories_pie = ['Slice 1','Slice 2','Slice 3','Slice 4']
#values_pie = ['15','30','25','30']
values_pie = [10,30,25,35]

plt.figure(figsize = (8,8))
plt.pie(values_pie,labels = categories_pie,autopct = '%1.1f%%',startangle = 90,colors = ['red','blue','orange','pink'])
plt.title('Pie Chart')
plt.show()
```

### Histogram

```python
#Generate random data for this histogram
data_hist = np.random.randn(1000) #Example random data

#plot histogram
plt.figure(figsize = (10,5))
plt.hist(data_hist,bins = 30,color = 'orange',edgecolor = 'black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

### Scatter plot

```python
#Scatter Plot

#Generate random data for the scatter plot
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

#Create scatterc plot
plt.figure(figsize = (8,8))
plt.scatter(x_scatter,y_scatter,color = 'green',marker = 'o')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
